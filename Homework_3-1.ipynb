Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
df.describe()
```

Поиск выбросов по столбцу Age:

```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['Age'].between(lower_bound, upper_bound, inclusive=True)].sort_values('Age')
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
```

По столбцу Age выбросом является значение 9, т.к. по описанию в данный столбец должны заносится только значения 1 или 2

Поиск выбросов по столбцу rectal temperature:

```python
q1 = df['rectal temperature'].quantile(0.25)
q3 = df['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal temperature')
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
```

Выбросы - значения, ниже, чем значение квантиль 25% = 37.8 и выше, чем квантиль 75% = 38.5, также выведены пропуски

Поиск выбросов по столбцу pulse:

```python
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
```

Выбросы - значения выше, чем квантиль 75% = 88, также выведены пропуски

Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
(df.isna().mean() * 100).round()
```

Пропуски столбца pain можно восстановить на основании статистики и группировки столбца surgery?, т.к. между уровнем боли лошади и хирургическим вмешательстве возможна взаимосвязь.

```python
df['pain'].fillna(df.groupby('surgery?')['pain'].transform('median'), inplace = True)
```

Пропуски столбца pulse можно восстановить по значению медианы, т.к. данная метрика = 64.0 отражает отклонение от нормы (30-40 ударов) и значение по медиане не так подвержено выбросам, как метрика mean.

```python
df['pulse'].fillna(df['pulse'].median(), inplace = True)
```

Пропуски столбца respiratory rate можно восстановить по значению медианы, т.к. данная метрика показывает отклонение от нормальной скорости 8-10 и значение по медиане не так подвержено выбросам, как метрика mean.

```python
df['respiratory rate'].fillna(df['respiratory rate'].median(), inplace = True)
```

Для столбца rectal temperature можно попробовать сгруппировать по столбцу outcome, для значений outcome 2,3 rectal temperature близка с min либо max значениям, что говорит о состоянии шока (осложнение заболевания).

```python
df['rectal temperature'].fillna(df.groupby('outcome')['rectal temperature'].transform('median'), inplace = True)
```

Столбец temperature of extremities можно восстановить через группировку по столбцу rectal temperature по медиане, т.к. между показателями существует прямая связь (чем выше rectal temperature, тем выше temperature of extremities). 

```python
df['temperature of extremities'].fillna(df.groupby('rectal temperature')['temperature of extremities'].transform('median'), inplace = True)
```
